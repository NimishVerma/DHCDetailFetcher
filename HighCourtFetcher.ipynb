```python
import requests

url = "http://delhihighcourt.nic.in/parse.asp"
advocate_name = "ravi" #can be full name or substring
#for given date 18 june 2018, for advocates with mayank in their name as substring
payload = "drpMonth=6&drpDay=18&drpYear=2018&criteria=SUBSTRING&advocateNameAsSunstring="+advocate_name+"&courtnoId=&Submit=Submit&advName=&judgeName="
headers = {
    'content-type': "application/x-www-form-urlencoded",
    'cache-control': "no-cache",
    'postman-token': "3650466b-002d-86c5-2293-3d1960c58dc4"
    }

response = requests.request("POST", url, data=payload, headers=headers)

print(response.text)
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(response.text)
```

```python

judge_list = soup.find_all(id='tcol2')
print(judge_list)
```

```python
judge_list=judge_list[1:]#remove header
```

```python
courtnumbers = soup.find_all('td',id="")
courtnumbers = courtnumbers[1:] #remove header coloumn
```

```python
for i in range(len(courtnumbers)):
    print (courtnumbers[i].text)
    judgestring = judge_list[i].text
    a=judgestring.split("HON'BLE")
    a = a[1:] #remove first comma
    for j in a:
        print ("HON'BLE" + j)
    
```
